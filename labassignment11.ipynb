### Part c
Create a visualization with 
* nine barplots, arranged in a 3x3 grid. 
* The barplots should refer to each of the nine categories of `region`, 
* and each barplot should be given a label that contains the name of the region.
* Within each barplot, list the categories of `relationship`,
* and display horizontal bars.

Use `seaborn`, `matplotlib`, and the `.plot()` method of a pandas dataframe as you see fit. 

[10 points]

````python
# --- Normalize and set ordered categorical for 'relationship' ---
relationship_order = ['strongly agree', 'agree', 'disagree', 'strongly disagree']

# Normalize text entries (lowercase, strip) and coerce to NaN if not in valid set
valid = set(relationship_order)
gss_clean['relationship'] = (
    gss_clean['relationship']
    .astype(str)
    .str.strip()
    .str.lower()
    .where(lambda s: s.isin(valid))
)

# Set ordered categorical dtype
gss_clean['relationship'] = pd.Categorical(
    gss_clean['relationship'],
    categories=relationship_order,
    ordered=True
)

# Remove unused categories to keep grouping behavior explicit
gss_clean['relationship'] = gss_clean['relationship'].cat.remove_unused_categories()

# --- Prepare list of up to 9 regions (sorted for reproducibility) ---
regions = sorted(gss_clean['region'].dropna().unique())[:9]

regions
````

````python
# --- Plotting: 3x3 grid of horizontal barplots ---
sns.set_style("whitegrid")
fig, axes = plt.subplots(3, 3, figsize=(15, 12), sharex=False, sharey=False)
axes = axes.flatten()

for ax, reg in zip(axes, regions):
    # Count responses for this region.
    # NOTE: set observed=True so groupby on categorical uses only observed categories (avoids FutureWarning)
    counts = (
        gss_clean[gss_clean['region'] == reg]
        .groupby('relationship', observed=True)
        .size()
        .reindex(relationship_order, fill_value=0)
    )

    # Plot horizontal bars using pandas plotting (barh)
    counts.plot(kind='barh', ax=ax, color='C0', legend=False)

    # Title and labels
    ax.set_title(str(reg), fontsize=12)
    ax.set_xlabel('Number of respondents')
    ax.set_ylabel('')  # relationship labels are on the y-axis

    # Annotate each bar with its count
    for p in ax.patches:
        width = int(p.get_width())
        if width > 0:
            ax.annotate(
                f'{width}',
                xy=(width, p.get_y() + p.get_height() / 2),
                xytext=(4, 0),
                textcoords='offset points',
                ha='left',
                va='center',
                fontsize=9
            )

# Hide any unused subplots if fewer than 9 regions
if len(regions) < 9:
    for ax in axes[len(regions):]:
        ax.axis('off')

plt.suptitle('Responses to: "A working mother can establish just as warm and secure a relationship" by Region', fontsize=14)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
````

````python
# --- Keep and rename relevant columns ---
gss = gss[['educ','prestg10','sei10']].rename({
    'prestg10':'job_prestige',
    'sei10':'socioeconomic_index'
}, axis=1)

# --- Convert education to numeric and drop missing rows ---
gss['educ'] = pd.to_numeric(gss['educ'], errors='coerce')
gss = gss.dropna(subset=['educ','job_prestige','socioeconomic_index'])

# --- Exclude education values less than 8 as instructed ---
gss = gss[gss['educ'] >= 8]

# --- Compute means by education level ---
means = (gss
         .groupby('educ', observed=True)[['job_prestige','socioeconomic_index']]
         .mean()
         .reset_index()
        )

means
````

````python
# --- Prepare data for seaborn (long form) ---
plot_df = means.melt(id_vars='educ',
                     value_vars=['job_prestige','socioeconomic_index'],
                     var_name='measure',
                     value_name='mean_value')

# Map nicer labels for the legend
label_map = {
    'job_prestige': 'Occupational prestige',
    'socioeconomic_index': 'Socioeconomic status'
}
plot_df['measure_label'] = plot_df['measure'].map(label_map)
````

````python
# --- Plot 2: Distribution of prestige quartiles by sex (proportion of each sex in each quartile) ---
prop_by_sex = g.groupby('sex')['prestige_q'].value_counts(normalize=True).unstack().fillna(0)
# Reorder columns to match quartile order
prop_by_sex = prop_by_sex[['Q1_low','Q2','Q3','Q4_high']]

ax2 = prop_by_sex.plot(kind='bar', stacked=True, figsize=(9,5), colormap='Pastel1')
ax2.set_title('Distribution of prestige quartiles by sex', fontsize=14)
ax2.set_xlabel('Sex', fontsize=12)
ax2.set_ylabel('Proportion of sex in each quartile', fontsize=12)
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))

# Annotate stacked bars with percent segments
for i, sex in enumerate(prop_by_sex.index):
    cum = 0.0
    for j, quart in enumerate(prop_by_sex.columns):
        val = prop_by_sex.loc[sex, quart]
        if val > 0.02:  # only label segments large enough to read
            ax2.text(i, cum + val/2, f'{val:.0%}', ha='center', va='center', fontsize=9)
        cum += val

plt.legend(title='Prestige quartile', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
````

Interpretation:

Flips the perspective to show, for each sex, how their workforce is distributed across quartiles. 

Men have a larger share in the top quartile and a smaller share in the bottom quartile compared with women, confirming the same sorting effect from the sex viewpoint. 

This chart is useful for reporting the proportionate contribution of sorting to the overall gap.
